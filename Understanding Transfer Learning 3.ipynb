## AlexNet as fixed feature extractor

```python
import torchvision
import torch.nn as nn

model = torchvision.models.alexnet(pretrained=True)
mod = list(model.classifier.children())
mod.pop()
model.classifier = nn.Sequential(*mod)
```

## Extract the features

```python
from torchvision import transforms
from torch.autograd import Variable

import torch

from load_data import ImageFilelist

transform = transforms.Compose(
    [transforms.Scale(size=(223,223)),
     transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

dataset = ImageFilelist(root="./data", flist="dataset.txt", transform=transform)
trainloader = torch.utils.data.DataLoader(dataset, batch_size=6, shuffle=False, num_workers=6)
dataiter = iter(trainloader)
images, labels = dataiter.next()

for i in range(6):
    output = model(Variable(images[i].view(1, 3, 223, 223)))
    print(output, torch.nonzero(output.data))
```
